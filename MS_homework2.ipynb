```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
np.random.seed(2060)

a = -4
teta = 8
var = 0.5 * teta - 1
sigma = np.sqrt(var)

n = 200
bins = 10

data = np.random.normal(a, sigma, n)
x_grid = np.linspace(np.min(data), np.max(data), 1_001, endpoint=True)

hist, bin_edges = np.histogram(data, density='True', bins=bins)
x_list = np.array([], dtype='float64')

for i in range(1, bin_edges.shape[0]):
    x_list = np.append(x_list, (bin_edges[i-1] + bin_edges[i]) / 2)

plt.hist(data, bins=bins, density=True, alpha=0.5, color='turquoise', label='histogram')
plt.plot(x_list, hist, '-o', color='lime', label='freq. polygon')
plt.plot(x_grid, norm.pdf(x_grid, a, sigma), color='gold', label='theoretical $f_{\\xi}(x)$')

plt.title('Histogram, polygon of frequencies and density function of distibution', y = 1.05)
plt.legend(loc='best')
plt.ylim([0, 1])
plt.grid(ls='-')

#plt.savefig(fname='task2_2a.pdf', format='pdf')

plt.show()
```
